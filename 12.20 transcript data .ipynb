## Formating the transcript data with HU treatment
Download the genome feature table from the website (ftp://ftp.ncbi.nlm.nih.gov/genomes/all/GCA/000/007/005/GCA_000007005.1_ASM700v1) 

```python
import pandas as pd
df_genome=pd.read_table('../data/GCA_000007005.1_ASM700v1_feature_table.txt')
df_HU_treatment =pd.read_excel('../data/transciptone.xlsx',sheet_name = "HU treatment of sulfolobus ")
```

```python
df_genome = df_genome.set_index(["# feature"],)
df_genome = df_genome.drop(index='gene')
```

```python
df_HU_treatment = df_HU_treatment.set_index(["START"],)
df_HU_treatment["ORF"] = ""
for start in df_HU_treatment.index:
    for i in range(len(df_genome)):
        if start == df_genome["start"].iloc[i]:
            df_HU_treatment["ORF"].at[start] = df_genome["locus_tag"].iloc[i]     
```

```python
df_HU_treatment.set_index("ORF")
```

## Formating the transcript data for different oxygen concentration

```python
df_oxygen_transcript =pd.read_excel('../data/transciptone.xlsx',sheet_name = "26vs1.5 hyb_2")
#drop out the one without SSO number
df_oxygen_transcript = df_oxygen_transcript.loc[df_oxygen_transcript['#ID_REF = '].str.contains('SSO') == True]
```

```python
df_oxygen_transcript["ORF"] = ""
df_oxygen_transcript['transcript'] = ""
df_oxygen_transcript["position"] = ""
for i in range (len(df_oxygen_transcript)):
    ORF = df_oxygen_transcript["#ID_REF = "].iloc[i].split("_")
    df_oxygen_transcript['ORF'].iloc[i] = ORF[0]
    position = ORF[1].split(" ")
    df_oxygen_transcript["position"].iloc[i] = position [0]
    df_oxygen_transcript["transcript"].iloc[i] = position[-1]   
```

## Adding the gene reaction rule to the model 

```python
# set the envoiroment
import cobra
from cobra import Model, Reaction, Metabolite
from cobra.flux_analysis import flux_variability_analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import escher
from escher import Builder
from utils import show_map
from utils.check_precursor_problem import check_precursor_problem
```

```python
#load the model
model = cobra.io.read_sbml_model('../models/Sulfolobus-solfataricus_D-glucose.xml')
```

```python
# load the excel file that contain the gene reaction rule information
df_gene_reaction_rule =pd.read_excel('../data/transciptone.xlsx',sheet_name = "gene_reaction_rule")
```

```python
# change the reaction_id and make it exactly like the one used in the model
for i in range(len(df_gene_reaction_rule)):
    df_gene_reaction_rule["reaction_id"].iat[i] = ((str(df_gene_reaction_rule["reaction_id"].iloc[i])[:-1]).strip())[1:-1]
# set the reaction id column as the index in order to map it to the model
df_gene_reaction_rule = df_gene_reaction_rule.set_index(["reaction_id"],)
```

```python
# add tge gene reaction rule to the model
reactions_ids = [rxn.id for rxn in model.reactions]
for rxn in df_gene_reaction_rule.index:
    if rxn in reactions_ids:
        if rxn != "oxp_redox_1.6.5.3_RXN0__5330":
            if 'spontaneous' not in str(df_gene_reaction_rule['genes'].loc[rxn]):
                c = df_gene_reaction_rule["genes"].loc[rxn].split(",")
                if len(c) > 1:
                    d = "SSO" + str(" "+ str(df_gene_reaction_rule['gene_reaction_rule'].loc[rxn]) +" SSO").join(c)
                    model.reactions.get_by_id(rxn).gene_reaction_rule = '( '+ d + ' )'
                if len(c) == 1:
                    model.reactions.get_by_id(rxn).gene_reaction_rule = '( SSO'+ df_gene_reaction_rule["genes"].loc[rxn] + ' )'
        else:
            model.reactions.get_by_id("oxp_redox_1.6.5.3_RXN0__5330").gene_reaction_rule  = '(( SSO0322 and SSO0323 and SSO0324 and SSO0325 and SSO0326 and SSO0327 and SSO0328 and SSO0329 and SSO0665 ) or (SSO1010 and SSO2960 and SSO3148 ))'
```
