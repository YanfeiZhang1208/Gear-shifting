```python
##Robustness analysis： in this method the flux through one reaction is varied and the 
## and the optimal objective value is calculated as a function of this flux. This 
## reveals how sensitive the objective is to a particular reaction. There are many insightful 
## combinations of reaction and objective that can be investigated with robustness analysis. 
## One fancy use is examination of the effects of nutrient uptake on growth rate. 
for i in range (-6,0):
    model.reactions.get_by_id("EX_glu_L_e").upper_bound = i
    model.reactions.get_by_id("EX_glu_L_e").lower_bound = i
    print (model.optimize())
```

```python
## Phenotypic phase plane analysis : when performing robustness analysis, one parameter is 
## varied and the network state is calculated. It is also possible to vary two parameters simultaneously and 
## plot the results as phenotypic phase plane. These plots can reveal the interaction
## between two reactions in interesting ways. These can be done in a manner similar to the calculations for 
## robustness analysis. Instead of using one for loop to vary one reaction, two nested 
## for loops are used to vary two reactions. In this case, use for loops to vary the bounds of the glucose exchange and oxygen 
## exchange reaction between 0  and -20. 
for i in range (0,20):
    for j in range (0,20):
        model.reactions.get_by_id("EX_glu_L_e").upper_bound = -i
        model.reactions.get_by_id("EX_glu_L_e").lower_bound = -i
        model.reactions.get_by_id("EX_co2_e").lower_bound = j
        model.reactions.get_by_id("EX_co2_e").lower_bound = j
        print (model.reactions.get_by_id("EX_glu_L_e").upper_bound)
        print (model.reactions.get_by_id("EX_co2_e").lower_bound)
        print (model.optimize())
```
